# Investigation into timings

To get more insight into how the synthesized models actually operate on the FPGA (with PYNQ and the overhead of datatrasnfers/memorycontrol/controlsignals), we did some preliminery investigation to find out how large the overhead is.

Our observations are that the execution time is wildly different from run to run, and not closely linked to nr of samples/batch size. 
This is non-conclussive, but our hypothesis is that the overhead we observe in the execution-time is closely dependent on the imperfect time-capturing dependent on the CPU influenced by scheduler/queue, and moving of data.

```python
import numpy as np
x_test_path = 'x_test.npy'
y_test_path = 'y_test.npy'
x_test = np.load(x_test_path)
y_test = np.load(y_test_path)
```

```python
import os
import time
from axi_master_driver import NeuralNetworkOverlay

timestamp = time.strftime("%Y%m%d_%H%M%S")

nr_samples = 10
iterations = 1000
timings = []

x_test_samples = x_test[:nr_samples]
y_test_samples = y_test[:nr_samples]
print(x_test_samples.shape,y_test_samples.shape)

bitfile = 'DUT/Training_AdaptiveHP_acc=0.7117_ebops=303_VU_DA_bitfile/system.bit'
overlay = NeuralNetworkOverlay(bitfile_name=bitfile, x_shape=x_test_samples.shape, y_shape=y_test_samples.shape, dtype=x_test_samples.dtype)

for i in range(iterations):
    # Do the prediction/run inference
    result = overlay.predict(x_test_samples, debug=False, profile=True, encode=np.float32, decode=np.float32)
    timings.append(result[1])

    
timing_results_path = f"timings_smp-{nr_samples}_it-{iterations}_{timestamp}.txt"
np.savetxt(timing_results_path,timings)
```

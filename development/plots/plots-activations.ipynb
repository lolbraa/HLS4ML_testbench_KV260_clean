# Activation function demonstration

Conda environment for this notebook: `conda env create`.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def s(x):
    return 1 / (1+np.exp(-x))
def r(x):
    return np.maximum(0, x)
```

```python
plotname = "Demonstration of Activation Functions"

x = np.linspace(-4,4,100)
y_s = s(x)
y_r = r(x)

fig, ax = plt.subplots(1, 1, figsize=(10, 6))

plt.plot(x, y_s, label="Sigmoid")
plt.plot(x, y_r, label="ReLU")

ax.set_ylim(-0.01,4)
ax.set_xlim(-4,4)
ax.legend()


ax.grid(True)
for spine in ("right", "top",
            "left", "bottom"
            ):
    ax.spines[spine].set_visible(False)

plt.suptitle(plotname, size = 'xx-large')#, weight = 'bold')
plt.tight_layout()
plt.show()

fig.savefig(f"export/{plotname.lower()}.pdf", format='pdf', bbox_inches="tight")
```

```python
def cross_entropy(x):
    return -np.log(x)
```

```python
print(cross_entropy(0.5))
```

```python
plotname = "Cross Entropy loss"

x = np.linspace(0,1,500)
y_loss = cross_entropy(x)

fig, ax = plt.subplots(1, 1, figsize=(10, 6))

plt.plot(x, y_loss, label="CE")

ax.set_ylim(0,4)
ax.set_xlim(0,1)
ax.legend()


ax.grid(True)
for spine in ("right", "top",
            "left", "bottom"
            ):
    ax.spines[spine].set_visible(False)

plt.suptitle(plotname, size = 'xx-large')#, weight = 'bold')
plt.tight_layout()
plt.show()

fig.savefig(f"export/{plotname.lower()}.pdf", format='pdf', bbox_inches="tight")
```
